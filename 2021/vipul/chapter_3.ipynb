```python
import json
import re
```

## Problem 20

```python
data = None
with open('enwiki-country.json') as infile:
    data = [json.loads(i) for i in infile.readlines()]
    
uk_idx = None
for i in range(len(data)):
    if data[i]['title'] == 'United Kingdom':
        uk_idx = i
        print(data[i]['text'][:20])
        break
```

## Problem 21

```python
def get_catergory(n):
    categs = []
    for line in data[n]['text'].split('\n'):
        if '[[Category:' in line:
            categs.append(line)
    return categs

def get_category2(n):
    categs = []
    for line in data[n]['text'].split('\n'):
        if re.search(r"^\[\[Category:.*\]\]$", line) is not None:
            categs.append(line)
            
    #shortest way
    x = re.findall(r"\[\[Category:.*\]\]", data[n]['text'])
    return x, categs


get_catergory(uk_idx)
get_category2(uk_idx)
```

## problem 22

```python
def get_category_names(n):
    pattern = r"^\[\[Category:(.*?)(?:\|.*)?\]\]$"
    p = re.compile(pattern, re.MULTILINE)
    x = p.findall(data[n]['text'])
    return x


get_category_names(uk_idx)
```

## problem 23

```python
from IPython.display import Markdown, display
def get_section(n):
    pattern = r'^(={2,})(.*?)\1$'
    p = re.compile(pattern, re.MULTILINE)
#     print(*data[n]['text'].split('\n'), sep='\n')
    x = p.findall(data[n]['text'])
    return x


for i in get_section(uk_idx):
    level = len(i[0]) - 1
    hsize = min([6, level])
    out =  ''.join(['#' for i in range(level)]) + ' ' + f'{level}. {i[1]}'
    display(Markdown(out))
```

## problem 24

```python
def get_ref(n):
    pattern = r'\[\[File:(.*?)\|'
    p = re.compile(pattern, re.MULTILINE)
    x = p.findall(data[n]['text'])
    return x
    
get_ref(uk_idx)
```

## problem 25

```python
def extract_infobox(n):
    infobox_pattern = r'\{\{Infobox.*?(.*?)^\}\}$'
    p = re.compile(infobox_pattern, re.MULTILINE | re.DOTALL)
    infobox_ = p.findall(data[n]['text'])[0]
    country_info = {}
#     print(infobox_)
    info_pattern = r'\|\s*(.+?)\s*=\s*(.+?)\s*\n'
    q = re.compile(info_pattern, re.MULTILINE)
    for i, j in q.findall(infobox_):
        country_info[i] = j
    return country_info

infobox = extract_infobox(uk_idx)
infobox['national_anthem']
# infobox
```

## problem 26

```python
def rm_emphasis(info):
    pattern = r'(\'{2,4}|\'{5,6})'
    p = re.compile(pattern)
    for i in info.keys():
        info[i] = p.sub('', info[i])
        
    
rm_emphasis(infobox)
print(infobox['national_anthem'])
print()
print(infobox['other_symbol_type'])
print()
print(infobox['image_map'])
# infobox
```

## problem 27

```python
# print(*data[uk_idx]['text'].split('\n'), sep='\n')
def rm_internal(info):
    pattern = r'\[\[(?:[^|]*?\|)??([^|]*?)\]\]'
    p = re.compile(pattern, re.MULTILINE)
    for i in info.keys():
        info[i] = p.sub(r'\1', info[i])
        
rm_internal(infobox)

print(infobox['other_symbol_type'])
print()
print(infobox['image_map'])
# infobox
```

## problem 28

```python
def rm_all(info):
    pattern1 = r'(?:url=)?https?://\S*?\s'
    pattern2 = r'<.+?>'
    pattern3 = r'\[\[File:([^|]*)\|([^|]*)\|([^|]*?)\]\]'
    pattern4 = r'\{\{(c|C)ite web(\s*)?([^|]*?\|)*?([^|]*?)\}\}'
    for i in info.keys():
        info[i] = re.sub(pattern1, '', info[i])
        info[i] = re.sub(pattern2, '', info[i])
        info[i] = re.sub(pattern4, '', info[i], re.IGNORECASE)
        info[i] = re.sub(pattern3, r'\1', info[i])
        
rm_all(infobox)
infobox['image_map']
```
